RQ2: Which suppliers deliver poor quality or have longest delays?

```python
df.groupby('Supplier name')[['Defect rates', 'Lead time']].mean()
```

```python
supplier_perf = (
    df.groupby('Supplier name')[['Defect rates', 'Lead time']]
    .mean()
    .sort_values(['Defect rates', 'Lead time'], ascending=[False, False])
)

supplier_perf
```

Suppliers with highest defect rates and longest lead times:

```python
print(supplier_perf.head())
```

```python
supplier_perf.plot(
    kind='scatter',
    x='Lead time',
    y='Defect rates',
    figsize=(8,5),
    title='Supplier Quality vs Lead Time'
)
plt.xlabel('Average Lead Time (days)')
plt.ylabel('Average Defect Rate')
plt.grid(True)
plt.show()
```

RQ3: How do logistics decisions (mode, route) influence cost performance?

```python
mode_costs = df.groupby('Transportation modes')['Costs'].mean().sort_values()
print("\nAverage cost per transportation mode:\n", mode_costs)
```

```python
route_costs = df.groupby('Routes')['Costs'].mean().sort_values()
print("\nAverage cost per route:\n", route_costs.head(10))
```

```python
mode_route_costs = (
    df.groupby(['Transportation modes', 'Routes'])['Costs']
    .mean()
    .reset_index()
    .sort_values('Costs', ascending=False)
)
print("\nTop 10 (mode, route) combinations with highest average cost:\n")
print(mode_route_costs.head(10))
```

```python
if 'Shipping times' in df.columns:
    corr = df[['Shipping times', 'Costs']].corr().iloc[0, 1]
    print(f"\nCorrelation between shipping time and cost: {corr:.2f}")

# Visualization: cost by transportation mode

mode_costs.plot(kind='bar', figsize=(7,4), title='Average Cost per Transportation Mode')
plt.ylabel('Average Cost')
plt.show()

# Visualization: scatter of cost vs shipping time (if available)
if 'Shipping times' in df.columns:
    df.plot(
        kind='scatter',
        x='Shipping times',
        y='Costs',
        c='blue',
        alpha=0.6,
        figsize=(7,5),
        title='Relationship Between Shipping Time and Cost'
    )
    plt.xlabel('Shipping Time')
    plt.ylabel('Cost')
    plt.grid(True)
    plt.show()
```

How can interactive charts speed up cycle time analysis?

Feature	Benefit for Cycle Time Analysis
Filtering and drilling down	You can filter by product type, supplier, route, or transportation mode instantly — no need to re-run code.
Hover tooltips	When hovering over data points, you see exact values (lead time, cost, defect rate, etc.), helping pinpoint anomalies.
Zooming and panning	Helps analyze specific time periods (e.g., seasonal peaks or disruptions).
Dynamic linking	Selecting a long lead time in one chart automatically highlights related suppliers or products in others.
Real-time updates	If linked to live data (e.g., via Grafana, Power BI, Plotly Dash), analysts can detect delays as soon as they happen.
Scenario comparison	You can adjust filters to compare cycle times across logistics routes or suppliers interactively.

```python
fig = px.scatter(
    df,
    x="Lead time",
    y="Costs",
    color="Transportation modes",
    size="Shipping times",
    hover_data=["Routes", "Supplier name", "Location"],
    title="Visual Prediction of High-Cost Transport Routes"
)
fig.show()
```

interpretation

Points higher on the y-axis → higher costs.

Larger bubbles → longer shipping times.

Color groups → different transportation modes (air, road, sea, etc.).

You can hover over outliers to identify which route/supplier is responsible for extreme costs.

```python
# Aggregate total demand by region and customer segment
demand_by_region = df.groupby(['Location', 'Customer demographics'])['Number of products sold'].sum().reset_index()

# Sort descending by demand
demand_by_region = demand_by_region.sort_values('Number of products sold', ascending=False)
print(demand_by_region.head(10))
```

Pivot Table for Heatmap Visualization

```python
# Pivot for heatmap
pivot_table = demand_by_region.pivot(index='Customer demographics', columns='Location', values='Number of products sold').fillna(0)

print(pivot_table)
```

Visualize Patterns with Heatmap

```python

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Demand by Customer Demographics and Region")
plt.ylabel("Customer Demographics")
plt.xlabel("Region")
plt.show()
```

Optional Interactive Version with Plotly

```python

fig = px.sunburst(
    df,
    path=['Location', 'Customer demographics'],
    values='Number of products sold',
    color='Number of products sold',
    color_continuous_scale='Viridis',
    title='Demand Distribution by Region and Customer Segment'
)
fig.show()
```
